```python
import nltk
import string

# Download required resources (first time only)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from collections import Counter
```

```python
text = """
Natural Language Processing is a branch of Artificial Intelligence.
It helps computers understand human language.
Students are learning NLP techniques.
"""

print("Original Text:")
print(text)
```

```python
text = text.lower()

print("\n1. Lowercase:")
print(text)
```

```python
# ==========================================
# 2. Tokenization
# ==========================================

tokens = word_tokenize(text)

print("\n2. Tokens:")
print(tokens)
```

```python
# ==========================================
# 3. Remove Punctuation
# ==========================================
tokens = [word for word in tokens if word not in string.punctuation]

print("\n3. After Removing Punctuation:")
print(tokens)
```

```python
# ==========================================
# 4. Stopword Removal
# ==========================================
stop_words = set(stopwords.words('english'))

filtered_words = [
    word for word in tokens
    if word not in stop_words
]

print("\n4. After Stopword Removal:")
print(filtered_words)
```

```python
# ==========================================
# 5. Stemming
# ==========================================

stemmer = PorterStemmer()

stemmed_words = [
    stemmer.stem(word)
    for word in filtered_words
]

print("\n5. Stemmed Words:")
print(stemmed_words)
```

```python
# ==========================================
# 6. Lemmatization
# ==========================================

lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word)
    for word in filtered_words
]

print("\n6. Lemmatized Words:")
print(lemmatized_words)
```

```python
# ==========================================
# 7. POS Tagging
# ==========================================

pos_tags = nltk.pos_tag(filtered_words)

print("\n7. POS Tags:")
print(pos_tags)
```

```python







# ==========================================
# 8. Word Frequency
# ==========================================

freq = Counter(filtered_words)

print("\n8. Word Frequency:")
for word, count in freq.items():
    print(word, ":", count)
```
